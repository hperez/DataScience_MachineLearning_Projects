## Bank Loan Classification

```python
# Import Libraries

# Standard libraries
import pandas as pd

# Custom class libraries
import random_forest_custom
```

```python
# Read dataset
input_file_path = '/Users/test/Documents/git_workspace/bank_loan_classification/UniversalBank.csv'
input_data = pd.read_csv(input_file_path)
```

```python
# Get insights on the data
# input_data.head(10)
# input_data.shape[1]
input_data.describe()
```

```python
# Create a class object
class_obj = random_forest_custom.RandomForest()
X_train, X_test, y_train, y_test = class_obj.data_preparation(input_data, 0.7)
```

```python
# Train model
model = class_obj.train_model(20, X_train, y_train)
```

```python
# Predict results
y_pred = class_obj.predict_results(model, X_test)
```

```python
# Evaluate results

# 1. Accuracy score
accuracy_score = class_obj.evaluate_results(y_test, y_pred)
print(accuracy_score)
```
