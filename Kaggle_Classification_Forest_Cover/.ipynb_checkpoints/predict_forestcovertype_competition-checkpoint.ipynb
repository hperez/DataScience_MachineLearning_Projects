### Understanding Support Vector Machine (Classification)

- Effective in high-dimensional space.
- 2 rules for SVM - Maximum classification and Best separation.
- **Kernel Transformation** - When the data is not linearly separable, SVM projects the data into a higher dimension to make it linearly separable. This is essentially known as Kernel Transformation - Kernel takes input features and creates linearly separable in higher dimension.
- Hyperplane is of (n-1) dimension where n is number of features (example, If data in 3 dimensions, a 2-d plane can separate it).
- **Maximum margin classifier** - SVM tries to maximise the margins on either side of the hyperplane.
- When using SVM, input data should be normalized so that all features are on the same scale.

#### Types of kernels:
1. **Linear** - When data is linear - Usually it's not the case.
2. **RBF/Gaussian** (Most commonly used) - Projects data into Gaussian distribution - points get higher, after which a plane can be used to separate them. *Prone to overfitting and captures noise (need to understand more what that means)*
3. **Polynomial** (Good results) - Uses base features to create polynomial features. Problem - Number of high dimension features increase exponentially, making it computationally expensive.  

RBF and Polynomial kernels are for non-linear hyper-planes.

#### Kernel trick  
Transforming data points to higher-dimension is costly. SVM doesn't actually does the transformation.  
It computes decision boundaries in terms of similarity measures in a higher-dimensional feature space without performing a transformation. Hence, called as kernel trick.

#### Hard/Soft SVM

Hard SVM - Constraint that each data point must lie on the correct side of the margin and there should be no misclassification.  
Hard SVM - Doesn't allow any misclassification to happen.  
Hard SVM - Works on linearly separable data. If non-separable/non-linear data, then no hyperplane returned.  

Soft SVM - To solve this issue, we use Soft SVM which relaxes the hard constraints of Hard SVM.  
Soft SVM - Uses Regularization parameter (C)

**Regularization parameter (C)** - How much misclassification we want to avoid. Penalizes misclassification.  
High for Hard SVM, low for Soft SVM.  
High C - Low margin around decision boundary - Less misclassification - Can lead to overfitting.  
Low C - High margin around decision boundary - More misclassifications allowed - Smoother and generic boundaries.

*High C value*: If increasing C value is not affecting the accuracy, that means no more training points left in violation (in margin or misclassified). Hence, it's a good idea to stop increasing the C value after accuracy stops increasing as higher C values takes more time in fitting.  

*Low C value*: Means more support vectors - may increase prediction time.  
**Hence, adjusting C value involves a trade-off between fitting and prediction time.**  

**Gamma** (Only for RBF)
Controls the distance of influence of a single data point. Inverse of the radius of influence (similarity) of data points.
- Low gamma value - large similarity radius (i.e. more nearby points can be considered similar).  
- High gamma values - points have to be really close to be considered similar - else treated as separate points - hence, SVM tries to fit each points - leading to complex decision boundary - high chances of overfitting.

**Using C and Gamma**
- For Linear kernel - Only C
- RBF - C and Gamma - If high gamma, C's effect is negligible - If low gamma, C's effect is same as in Linear kernel.
- Typical values:  
0.0001 < gamma < 10  
0.1 < C < 100
